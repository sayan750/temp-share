**Outliers are data points that significantly deviate from the overall pattern of the data. They can impact my analysis in the following ways:-**

- Skewness: Outliers can cause the distribution to be skewed towards the extreme values. This skewness can affect statistical measures such as the mean, making it less representative of the central tendency.

- Influence on statistical tests: Outliers can distort statistical tests that assume normality or equal variances. If your analysis involves hypothesis testing or regression, outliers can affect the results and lead to incorrect conclusions.

- Model performance: If you're building a predictive model based on flight prices, outliers can influence the model's performance. They may introduce noise or bias the model towards predicting extreme values.

As a data scientist analyzing the Flight Price dataset to identify factors that affect flight prices, there are several features you can consider. Here are some key features that may impact flight prices:

1. Departure and Arrival Cities: The cities involved in the flight route can influence prices. Different cities may have varying demand, competition, and operating costs, leading to price variations.

2. Flight Distance: The distance between the departure and arrival cities can affect prices. Longer flights may require more fuel and incur higher operational costs, which can be reflected in the ticket prices.

3. Flight Class: The class or cabin type (economy, business, first class) can significantly impact prices. Higher-class tickets generally come with more amenities and services, leading to higher costs.

4. Seasonality: The time of year or season can influence flight prices. Peak travel seasons, holidays, and events can lead to increased demand and subsequently higher prices.

5. Advance Booking: The time between booking a flight and the departure date can affect prices. Airlines often use yield management techniques, where prices increase as the departure date approaches and seat availability decreases.

6. Airline Carrier: The airline operating the flight can impact prices. Different carriers may have varying pricing strategies, market positions, and service offerings, which can result in price variations.

7. Flight Timing: The time of day and day of the week can influence prices. Popular flight times or days may have higher demand, resulting in increased prices.
